## Encontrando Outliers

```python
#eliminar elementos fora dos padroes, primeiro carregamos os dados e aplicamos a media dos valores
data = pd.read_table('fruit_data_with_colors_miss.txt',na_values=['.','?'])
data = data.fillna(data.mean())
```

```python
macas = data[data['fruit_name'] == 'apple']
```

```python
macas.describe()
```

```python
#iremos utilizar tecnica de encontrar outliers a partir do desvio padrao
est = macas['mass'].describe() #armazenando as estatisticas da coluna massa das maças
```

```python
#as maças que forem retornadas por esse comando elas devem ser consideradas outliers, acima da media
macas[(macas['mass'] > est['mean'] + (est['std']) * 2)]
```

```python
#as maças que forem retornadas por esse comando elas devem ser consideradas outliers, abaixo da media
macas[(macas['mass'] < est['mean'] - (est['std']) * 2)]
```
